```python
from music21 import *
file_path = "./phony_a.mid"

### MIDI読み込み
s = converter.parse(file_path)
### 表示
display(s.analyze('key'))
display(analysis.pitchAnalysis.pitchAttributeCount(s, 'name'))
s[0].plot('pianoroll')
```

```python
file_path = "./phony_b.mid"

s = converter.parse(file_path)

display(s.analyze('key'))
display(analysis.pitchAnalysis.pitchAttributeCount(s, 'name'))
s[0].plot('pianoroll')
```
